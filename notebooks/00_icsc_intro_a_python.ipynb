### `for`

for nos permite recorrer ciertas estructuras y ejecutar instrucciones para cada elemento de ellas.

La primer estructura que veremos es `range`, que nos genera números para un rango determinado

````python
for i in range(10):
    print(i)
````

Puede que por algún motivo no querramos seguir en el `for`, en tal caso podemos usar el comando `break`

````python
for i in range(10):
    if i == 7:
        break
    print(i)
````

O queremos `saltar` una de las iteraciones: usamos el comando `continue`

````python
for i in range(10):
    if i % 2 == 0: # si es un número par, salteo
        continue
    print(i)
````

## Diccionarios

Los diccionarios (también conocidos hashes) son también pan y manteca de python. 

Un diccionario, esencialmente, guarda objetos pero en vez de estar indexados por posición como en una lista, los indexa por *claves*, que pueden ser de *casi* cualquier tipo.

En pocas palabras, convierte claves en valores



````python
# Diccionario de nombre de animales a qué animal son
# el 2 es por "to" 
animal2especie = {
    "Chatrán": "gato",
    "Flipper": "delfín",
    "Coyote": "coyote",
    "Topo Gigio": "topo",
    "Winnie the Pooh": "oso",
}

animal2especie["Winnie the Pooh"]
````

Podemos pedir las claves de un diccionario con `keys`

````python
animal2especie.keys()
````

Y los valores con `values`

````python
animal2especie.values()
````

Podemos iterar un diccionario sobre `items`, lo que nos devuelve un par (clave, valor)

````python
for k in animal2especie.items():
    print(k)
````

Una forma más elegante de usar esto es "desarmar" la tupla que nos devuelve `items`, reasignándola a dos variables `animal`, `especie`

````python
for (animal, especie) in animal2especie.items():
    print(f"{animal} es un {especie}")
````

¿Qué pasa si le pedimos un animal que no está en nuestra colección?

````python
animal2especie["Splinter"]
````

Tira una excepción! Podemos hacer dos cosas:
1. Catchear la excepción con un try-catch

````python
try:
    especie = animal2especie["Splinter"] 
    # Acá salta excepción, el print de abajo no se ejecuta
    print(especie) 
except KeyError: 
    # Viene para acá
    print("Hubo un error, el maestro Splinter no está en nuestra bbdd")
````

También podemos usar el método .get, especificando un valor por defecto

````python
animal2especie.get("Splinter", "animal")
````

## Clases y objetos

Hasta ahora dijimos que todo en `python` es un objeto. ¿Qué es un objeto, para `python`? Cualquier entidad que podamos manipular, que tenga un estado y un comportamiento asociado*. 

Por ejemplo, la lista de animales que armamos, el diccionario de animales a especie..la función cuadrado (!). Todo son objetos. 

Los programas que vamos a ver casi siempre se componen de llamar a "funciones" de un objeto. Por ejemplo

```python
a = []

a.append(1) # append es una función del objeto lista
```

A esta función `append` que llamamos sobre un objeto de tipo lista le llamamos **método**. 


````python
a = [] # creo un objeto de tipo lista, vacío

a.append(1) # llamo al método append del objeto a
a.append(2) # llamo al método append del objeto a nuevamente

a
````

````python
d = {
    "a":1,
    "b":2,
}

d.keys() # Acá estoy llamando el método .keys de d
````




Ahora, ¿cómo creamos los objetos? ¿cómo definimos su comportamiento? ¡con clases! La clase es, en pocas palabras, un objeto dedicado a crear otros objetos y definir su comportamiento. Podemos pensarlo como que la clase es la "idea" que tenemos de algo, y el objeto es la materialización de ello (para ponerlo en términos platónicos). 

Una clase vendría a ser una plantilla/template de cómo se comportan los objetos que creamos a partir de ella. Por ejemplo, podríamos definir la clase `Animal` y a partir de ella crear objetos `perro`, `gato`, `pato`, etc.

Nota: La teoría es un poco más compleja que esto, pero para nuestros fines está bien

### Atributos y métodos

Los objetos tienen dos elementos básicos:

1. Atributos: variables asignadas a cada objeto
2. Métodos: funciones "propias" de ese objeto que pueden ser llamadas



Para definir un a clase, tenemos que hacer un par de cosas:

1. Usar el keyword class junto a un nombre que defina nuestra nueva clase
2. Definir un método `__init__`, que reciba los parámetros de inicialización de ese objeto. Este método se llama **constructor**
3. Definir los métodos que acepta 

Los métodos son funciones que definidos dentro de una clase que reciben un parámetro `self` en primer lugar, que es una variable que apunta al objeto que es referenciado.

````python
class Persona:
    def __init__(self, nombre):
        self.nombre = nombre

    def decir_nombre(self):
        print(f"Me llamo {self.nombre}")

    def jugar_al(self, juego):
        print(f"{self.nombre} está jugando al {juego}")
````

Para crear un objeto `Persona`, llamamos a la clase con los parámetros de inicialización

````python
juan = Persona("Juan")
pedro = Persona("Pedro")
jose = Persona("José")
````

Podemos hurgar en sus atributos

````python
juan.nombre
````

````python
pedro.nombre
````

Podemos llamar a sus métodos

````python
juan.decir_nombre()
````

````python
juan.jugar_al("Truco")
````

## Funciones de alto orden y listas por comprensión

Esto de poder manipular funciones como objetos cualesquiera nos permite hacer algunas cosas interesantes en python: por ejemplo, la función `map`

¿Qué hace `map(f, lista)`? Devuelve una nueva lista que resulta de la aplicación de f a cada elemento de la lista

````python
cuadrado = lambda x: x*x
map(cuadrado, [1, 2, 3, 4])
````

Nos devolvió un `map` 🤔. Esto es por temas de eficiencia, muchas veces nos devuelve python un "iterador" sobre una lista. Si queremos ver el resultado, aplicamos `list` sobre esto

````python
list(map(cuadrado, [1, 2, 3, 4]))
````

Esto que hicimos es equivalente a usar una lista por comprensión, que tienen la forma

`[f(x) for x in lista]`

donde f es una función

Por ejemplo

````python
[cuadrado(x) for x in [1, 2, 3, 4]]
````

Podemos filtrar algunos elementos de la lista: por ejemplo, sólo los cuadrados de números menores a 100 que no sean divisibles por 2 o por 3

````python
[cuadrado(x) for x in range(100) if (x % 2 != 0)]
````

Una idea similar a la de listas por comprensión: diccionarios por comprensión

````python
{k:k**2 for k in range(20)}
````
